````python
from sklearn.base import BaseEstimator, TransformerMixin

class CustomMedianImputer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.medians = {}

    def fit(self, X, y=None):
        for category in X['Outcome'].unique():
            self.medians[category] = X[X['Outcome'] == category].median()
        return self

    def transform(self, X):
        for index, row in X.iterrows():
            for category in self.medians.keys():
                X.loc[index, X.columns[X.columns != 'Outcome']] = X.loc[index, X.columns[X.columns != 'Outcome']].fillna(self.medians[category])
        return X
````

````python
imputer = CustomMedianImputer()

diabetes_imputed = imputer.fit_transform(train_set)
````
